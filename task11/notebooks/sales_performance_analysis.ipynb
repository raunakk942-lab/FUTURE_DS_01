# Sales Performance & Business Insights

Prepared for Meridian Home Goods, a fictional local retail shop. This notebook mirrors the production pipeline used to clean the data, compute business metrics, and generate the client-ready report and dashboard.

## Workflow

1. Load or generate the raw sales extract.
2. Clean the dataset and document every transformation.
3. Build revenue, product, category, regional, and growth-opportunity summaries.
4. Export the charts and report artifacts used by the dashboard and PDF report.

```python
import sys
from pathlib import Path

import pandas as pd

PROJECT_ROOT = Path.cwd().resolve().parents[0] if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from scripts.sales_analysis import (
    ASSETS_DIR,
    PROCESSED_DIR,
    RAW_DIR,
    REPORTS_DIR,
    build_report_markdown,
    build_report_pdf,
    clean_sales_data,
    ensure_project_directories,
    export_analysis_bundle,
    export_data_notes,
    export_summary_artifacts,
    generate_synthetic_sales_data,
    save_dataframe,
)

ensure_project_directories()
export_data_notes(REPORTS_DIR / 'data_notes.md')
```

```python
raw_path = RAW_DIR / 'sales_raw.csv'
if raw_path.exists():
    raw_df = pd.read_csv(raw_path)
else:
    raw_df = generate_synthetic_sales_data()
    save_dataframe(raw_df, raw_path)

cleaning_result = clean_sales_data(raw_df)
cleaned_df = cleaning_result.dataframe
cleaned_df.head()
```

```python
analysis_bundle = export_analysis_bundle(cleaned_df, ASSETS_DIR)
analysis_bundle['cleaning_log'] = cleaning_result.log_lines
export_summary_artifacts(cleaned_df, analysis_bundle)
build_report_markdown(analysis_bundle, REPORTS_DIR / 'client_report.md')
build_report_pdf(analysis_bundle, REPORTS_DIR / 'client_report.pdf')

analysis_bundle['kpis']
```
